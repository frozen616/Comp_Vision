```python
import numpy as np
```

```python

def sig(z):
    return 1/(1+np.exp(-z))

def relu(z):
    return max(0,z)

def frange(start, stop, step=1.0):
    ''' "range()" like function which accept float type''' 
    i = start
    while i < stop:
        yield i
        i += step
```

```python
inputs = [1.0,5.0,9.0]



ihw = list(frange(0.01,0.12,0.01))

hbias = list(frange(0.13,0.17,0.01))

ohw = list(frange(0.17,0.25,0.01))

obias = [0.24, 0.25]

hidden_layer_1 = inputs[0]*ihw[0] + inputs[1]*ihw[4] + inputs[2]*ihw[8] + hbias[0]
hidden_layer_2 = inputs[0]*ihw[1] + inputs[1]*ihw[5] + inputs[2]*ihw[9] + hbias[1]
hidden_layer_3 = inputs[0]*ihw[2] + inputs[1]*ihw[6] + inputs[2]*ihw[10] + hbias[2]
hidden_layer_4 = inputs[0]*ihw[3] + inputs[1]*ihw[7] + inputs[2]*ihw[11] + hbias[3]
```

```python

#sigmoid
hidden_layer_1_sig = sig(hidden_layer_1)
hidden_layer_2_sig = sig(hidden_layer_2)
hidden_layer_3_sig = sig(hidden_layer_3)
hidden_layer_4_sig = sig(hidden_layer_4)

print(hidden_layer_1_sig, hidden_layer_2_sig, hidden_layer_3_sig, hidden_layer_4_sig)
```

```python
#relu
hidden_layer_1_relu = relu(hidden_layer_1)
hidden_layer_2_relu = relu(hidden_layer_2)
hidden_layer_3_relu = relu(hidden_layer_3)
hidden_layer_4_relu = relu(hidden_layer_4)

print(hidden_layer_1_relu, hidden_layer_2_relu, hidden_layer_3_relu, hidden_layer_4_relu)
```

```python
ol1_sig = hidden_layer_1_sig*ohw[0] + hidden_layer_2_sig*ohw[2] + hidden_layer_3_sig*ohw[4] + hidden_layer_4_sig*ohw[6] + obias[0]
ol2_sig = hidden_layer_1_sig*ohw[1] + hidden_layer_2_sig*ohw[3] + hidden_layer_3_sig*ohw[5] + hidden_layer_4_sig*ohw[7] + obias[1]

ol1_relu = hidden_layer_1_relu*ohw[0] + hidden_layer_2_relu*ohw[2] + hidden_layer_3_relu*ohw[4] + hidden_layer_4_relu*ohw[6] + obias[0]
ol2_relu = hidden_layer_1_relu*ohw[1] + hidden_layer_2_relu*ohw[3] + hidden_layer_3_relu*ohw[5] + hidden_layer_4_relu*ohw[7] + obias[1]


print (ol1_sig, ol2_sig, ol1_relu, ol2_relu)
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(dpi=80)
ax = fig.add_subplot(1,1,1)
table_data=[
    ["   ", 'Hidden Layer',' ', ' ',' ','Output Layer',' ' ],
    [" ",'Node0','Node1','Node2','Node3','Node0','Node1'],
    ['Sigmoid',hidden_layer_1_sig,hidden_layer_2_sig,hidden_layer_3_sig,hidden_layer_4_sig,ol1_sig,ol2_sig],
    ['ReLU',hidden_layer_1_relu,hidden_layer_2_relu,hidden_layer_3_relu,hidden_layer_4_relu,ol1_relu,ol2_relu],
]
table = ax.table(cellText=table_data, loc='center')
table.set_fontsize(20)
table.scale(4,5)
ax.axis('off')
plt.show()
```
